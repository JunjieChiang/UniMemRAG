```python
import os
os.environ.pop("http_proxy", None)
os.environ.pop("https_proxy", None)
os.environ["NO_PROXY"] = "localhost,127.0.0.1"

import pandas as pd

test_path = "../benchmark/encyclopedic_vqa/test.csv"
test_df = pd.read_csv(test_path)

questions = test_df[["question", "question_original", "question_type", "answer", "evidence", "evidence_section_id", "evidence_section_title", "dataset_image_ids", "dataset_name", "dataset_category_id", "wikipedia_url", "wikipedia_title", "encyclopedic_vqa_split", "wikipedia_url_used_in_train"]].copy()

split_columns = ["answer", "evidence_section_title", "evidence"]

def split_pipe_separated(value):
    if pd.isna(value):
        return []
    if isinstance(value, str):
        return [part.strip() for part in value.split("|")]
    return [value]

for column in split_columns:
    questions[column] = questions[column].apply(split_pipe_separated)

records = questions.to_dict("records")
```
